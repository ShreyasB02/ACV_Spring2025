```python
#Please load the github repo https://github.com/yformer/EfficientSAM

import os
import cv2
import numpy as np
import torch
import matplotlib.pyplot as plt
from PIL import Image
from torchvision import transforms
from efficient_sam.build_efficient_sam import build_efficient_sam_vitt, build_efficient_sam_vits
from matplotlib.patches import Rectangle


def clear_memory():
    import gc
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()


def apply_gaussian_blur(image_np, sigma):
    if sigma == 0:
        return image_np
    kernel_size = int(6 * sigma + 1)
    if kernel_size % 2 == 0:
        kernel_size += 1
    return cv2.GaussianBlur(image_np, (kernel_size, kernel_size), sigma)


def load_model_for_visualization(model_name):
    if model_name == "Ti":
        model = build_efficient_sam_vitt()
    else:
        import zipfile
        zip_path = "weights/efficient_sam_vits.pt.zip"
        if os.path.exists(zip_path) and not os.path.exists("weights/efficient_sam_vits.pt"):
            with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                zip_ref.extractall("weights")
        model = build_efficient_sam_vits()

    if torch.cuda.is_available():
        model = model.cuda()
    model.eval()
    return model


def generate_mask_for_visualization(model, image_tensor, input_points, input_labels):
    with torch.no_grad():
        predicted_logits, predicted_iou = model(image_tensor[None, ...], input_points, input_labels)
        sorted_ids = torch.argsort(predicted_iou, dim=-1, descending=True)
        predicted_logits = torch.take_along_dim(predicted_logits, sorted_ids[..., None, None], dim=2)
        mask = torch.ge(predicted_logits[0, 0, 0, :, :], 0).cpu().numpy()
        iou_score = float(predicted_iou[0, 0, 0].cpu().numpy())
        del predicted_logits, predicted_iou, sorted_ids
        clear_memory()
        return mask, iou_score


def generate_blur_with_auto_prompts(image_paths, blur_levels=[0, 2, 5, 7, 10], max_size=600):
    os.makedirs("Ti_box", exist_ok=True)
    os.makedirs("Ti_point", exist_ok=True)
    os.makedirs("S_box", exist_ok=True)
    os.makedirs("S_point", exist_ok=True)

    iou_plot_data = {
        "Ti_box": [],
        "Ti_point": [],
        "S_box": [],
        "S_point": []
    }

    def generate_auto_prompts(img_np):
        gray = cv2.cvtColor(img_np, cv2.COLOR_RGB2GRAY)
        _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_OTSU + cv2.THRESH_BINARY)
        contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        if not contours:
            return [[300, 300], [400, 400]], [250, 250, 600, 600]
        largest = max(contours, key=cv2.contourArea)
        x, y, w, h = cv2.boundingRect(largest)
        box = [x, y, x + w, y + h]
        cx, cy = x + w // 2, y + h // 2
        point_prompts = [[cx, cy], [x + 10, y + 10]]
        return point_prompts, box

    configs = [
        ("Ti", "points", "Ti_point"),
        ("Ti", "box", "Ti_box"),
        ("S", "points", "S_point"),
        ("S", "box", "S_box")
    ]

    for image_path in image_paths:
        image_name = os.path.splitext(os.path.basename(image_path))[0]
        print(f"\n Image: {image_name}")

        image = Image.open(image_path).convert("RGB")
        image_np = np.array(image)
        h, w = image_np.shape[:2]
        scale = min(1.0, max_size / max(h, w))
        if scale < 1.0:
            image_np = cv2.resize(image_np, (int(w * scale), int(h * scale)))

        point_prompts, box_prompts = generate_auto_prompts(image_np)

        for model_name, prompt_type, folder in configs:
            model = load_model_for_visualization(model_name)

            if prompt_type == "points":
                coords = point_prompts
                labels = [[1] * len(coords)]
            else:
                coords = [[box_prompts[0], box_prompts[1]], [box_prompts[2], box_prompts[3]]]
                labels = [[2, 3]]

            input_points = torch.tensor([[coords]]).float()
            input_labels = torch.tensor([labels])
            if torch.cuda.is_available():
                input_points = input_points.cuda()
                input_labels = input_labels.cuda()

            for sigma in blur_levels:
                blur_img = apply_gaussian_blur(image_np, sigma)
                blur_out_path = f"{folder}/{image_name}_sigma_{sigma}_input.png"
                Image.fromarray(blur_img).save(blur_out_path)
                image_tensor = transforms.ToTensor()(blur_img)
                if torch.cuda.is_available():
                    image_tensor = image_tensor.cuda()

                mask, iou = generate_mask_for_visualization(model, image_tensor, input_points, input_labels)
                config_key = folder
                print(f" {config_key} | {image_name} | Ïƒ={sigma} â†’ IoU = {iou:.3f}")
                iou_plot_data[config_key].append((sigma, iou))

                overlay = np.zeros((*mask.shape, 4))
                overlay[mask] = [1, 0, 0, 0.4]

                plt.figure(figsize=(6, 6))
                plt.imshow(blur_img)
                plt.imshow(overlay)

                if prompt_type == "points":
                    for i, (x, y) in enumerate(coords):
                        plt.plot(x, y, 'o', color='lime', markersize=8, markeredgecolor='white', markeredgewidth=2)
                        plt.text(x + 10, y - 10, f'P{i+1}', color='white', fontsize=10, fontweight='bold')

                if prompt_type == "box":
                    x1, y1, x2, y2 = coords[0][0], coords[0][1], coords[1][0], coords[1][1]
                    rect = Rectangle((x1, y1), x2 - x1, y2 - y1, linewidth=2.5, edgecolor='yellow', facecolor='none')
                    plt.gca().add_patch(rect)
                    plt.text(x1, y1 - 10, 'BOX', color='yellow', fontsize=10, fontweight='bold')

                plt.title(f"{image_name} Ïƒ={sigma} IoU={iou:.3f}")
                plt.axis("off")
                out_path = f"{folder}/{image_name}_sigma_{sigma}.png"
                plt.savefig(out_path, bbox_inches='tight', dpi=150)
                plt.close()

                del image_tensor
                clear_memory()

            del model
            clear_memory()

    # Plot final IoU degradation graph
    plt.figure(figsize=(10, 6))
    for config, values in iou_plot_data.items():
        values = sorted(values, key=lambda x: x[0])
        sigmas = [x[0] for x in values]
        ious = [x[1] for x in values]
        plt.plot(sigmas, ious, marker='o', label=config)

    plt.title("IoU vs Gaussian Blur Level (Ïƒ)")
    plt.xlabel("Blur Ïƒ")
    plt.ylabel("IoU")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.savefig("iou_vs_sigma_plot.png", dpi=150)
    plt.show()
    print("\nIoU degradation plot to 'iou_vs_sigma_plot.png'")



image_list = ["/content/000000116825.jpg", "/content/000000125952.jpg", "/content/000000134722.jpg", "/content/000000371472.jpg", "/content/000000395701.jpg"]
generate_blur_with_auto_prompts(image_list)
```
